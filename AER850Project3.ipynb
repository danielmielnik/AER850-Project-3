```python
from ultralytics import YOLO
import cv2
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

dataset = '/content/drive/MyDrive/datasets/data.yaml'


# Load a model
model = YOLO("yolov8n.pt")

# Train the model with cuda
results = model.train(data=dataset, epochs=120, batch=8, imgsz=992, device='cuda')
```

```python
model.val()
```

```python
results = model.predict(source="/content/drive/MyDrive/datasets/evaluation", save=True)
```

```python
# Iterate through results and adjust label font size
i = 0
for r in results:
    i += 1
    im_array = r.plot(line_width=5, font_size=1)
    im = Image.fromarray(im_array[..., ::-1])
    im.show()
    im.save(f"/content/runs/detect/image_{i}.jpg")
    plt.imshow(im)
    plt.axis('off')
    plt.show()
```
