```python
from ultralytics import YOLO
import cv2
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

dataset = '/content/drive/MyDrive/datasets/data.yaml'


# Load a model
model = YOLO("yolov8n.pt")  # load a pretrained model (recommended for training)

# Train the model with MPS
results = model.train(data=dataset, epochs=200, batch=4, imgsz=1032, device='cuda')
```

```python
model.val()
```

```python
results = model.predict(source="/content/drive/MyDrive/datasets/evaluation", save=True)

# Iterate through results and adjust label font size
for r in results:
    im_array = r.plot(line_width=2, font_size=1) # Adjust font_size as needed
    im = Image.fromarray(im_array[..., ::-1]) # Convert to PIL image
    # Display or save the image with smaller labels
    im.show() # Or save using im.save("path/to/image.jpg")
    # Or use matplotlib to display
    plt.imshow(im)
    plt.axis('off')
    plt.show()
```
